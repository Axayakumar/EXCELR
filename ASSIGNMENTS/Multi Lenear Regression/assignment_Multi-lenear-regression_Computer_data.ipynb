```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
# necessary dependencies for plotting
import matplotlib.pyplot as plt #python library for plot and graphs
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# Importing the dataset
datasets = pd.read_csv('C:/Users/Adinath/Desktop/Data_Science/Assignments/Multi Lenear regression/Computer_data.csv')
datasets.head()
```

```python
datasets['premium'] = datasets['premium'].eq('yes').astype(int)
datasets['multi'] = datasets['multi'].eq('yes').astype(int)
datasets['cd'] = datasets['cd'].eq('yes').astype(int)
datasets
```

```python
y = datasets.price.values
X = datasets.drop(['price'], axis = 1)
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.3)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
lr=linear_model.LinearRegression()
model=lr.fit(X_train, y_train)
```

```python
print(model.score(X_test, y_test))
predictions=model.predict(X_test)
print(mean_squared_error(y_test, predictions))
```

```python
Actual_value=y_test
plt.scatter(predictions, Actual_value, color='b')
plt.xlabel('predictions price')
plt.ylabel('Actual')
plt.title('LinearRegression')
plt.show()
```
